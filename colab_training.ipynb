```python
import base64
from IPython.display import HTML

def download_b64(filename):
    with open(filename, 'rb') as f:
        data = f.read()
    
    # Kodowanie do base64
    b64 = base64.b64encode(data).decode()
    
    # Tworzenie linku HTML
    payload = f"data:application/zip;base64,{b64}"
    html = f'''
        <a download="{filename}" href="{payload}" target="_blank">
            <button style="padding: 10px 20px; background-color: #4CAF50; color: white; border: none; cursor: pointer; border-radius: 5px;">
                Kliknij tutaj, aby pobra {filename} (Base64)
            </button>
        </a>
        <p style="color: gray; font-size: 12px; margin-top: 5px;">
            Uwaga: Przy duych plikach (jak Twj zip ~80MB) przegldarka moe chwil "pomyle".
        </p>
    '''
    return HTML(html)

# Wywoanie dla Twojego pliku
download_b64('run_output.zip')
```
